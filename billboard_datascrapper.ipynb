```python
import billboard
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
from dateutil.relativedelta import relativedelta
```

```python
# title – The title of the track.
# artist – The name of the artist, as formatted on Billboard.com.
# peakPos – The track's peak position on the chart as of the chart date, as an int (or None if the chart does not include this information).
# lastPos – The track's position on the previous week's chart, as an int (or None if the chart does not include this information). This value is 0 if the track was not on the previous week's chart.
# weeks – The number of weeks the track has been or was on the chart, including future dates (up until the present time).
# rank – The track's current position on the chart.
# isNew – Whether the track is new to the chart.
```

```python
music_df = pd.DataFrame(columns=['title','artist','peakPos','lastPos','weeks','rank','isNew'])

end_year = 2022

# Create a datetime object for the first date in the range
current_date = datetime.date(1958, 8, 4)

# Loop through each week from the start date to the end date
while current_date.year <= end_year:
    print(current_date)
    chart=billboard.ChartData('hot-100', current_date)
    for row in chart:
        new_row={'title':row.title,'artist':row.artist, 'peakPos':row.peakPos, 'lastPos': row.lastPos, 'weeks': row.weeks, 'rank': row.rank, 'isNew': row.isNew, 'date':current_date}
        music_df=music_df.append(new_row, ignore_index=True)
    current_date += relativedelta(months=1)
```

```python
music_df.to_csv('billboarddata.csv')
```
