```python
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
```

```python
music=pd.read_csv("merged_data.csv", index_col=0)
music["year"]=music["date"].apply(lambda x: datetime.date.fromisoformat(x).year)
s=music["year"].value_counts()
print(s)
sns.relplot(x=s.index, y=s.values)
```
