```python
music_df = pd.DataFrame(columns=['year', 'track', 'artist'])
for y in range(1960,2020,2):
    chart = billboard.ChartData('hot-100', '{year:d}-01-01'.format(year=y))
    for rank in range(0,50):
        new_row = {'year': y, 'track': chart[rank].title, 'artist': chart[rank].artist, 'rank':rank+1}
        music_df=music_df.append(new_row, ignore_index=True)
        print(new_row)
```

```python
print(music_df)
```

```python
def find_uri(row):
    response=sp.search("artist:{} track:{}".format(row["artist"],row["track"]), limit=1, type='track')
    items = response['tracks']['items']
    if(len(items)>0):
        uri=items[0]['uri']
    else:
        response=sp.search("track:{}".format(row["track"]), limit=1, type='track')
        items = response['tracks']['items']
        if(len(items)>0):
            uri=items[0]['uri']
        else:
            uri=None
    return uri

music_df['track_uri'] = music_df.apply(lambda row: find_uri(row), axis=1)
clean_music_df=music_df.dropna()
uri_list=clean_music_df['track_uri'].tolist()
```

```python
print(len(uri_list))
```

```python
def find_tempo(row):
    audio_features=sp.audio_features(tracks=[row['track_uri']])
    return audio_features[0]['tempo']
clean_music_df['tempo'] = clean_music_df.apply(lambda row: find_tempo(row), axis=1)
newseries=clean_music_df.groupby('year')['tempo'].mean()
print(newseries)
```
